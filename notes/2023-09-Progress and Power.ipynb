```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Adding nodes for technological advancements
technological_advancements = ["Steam Engine", "Spinning Jenny", "Telegraph"]
G.add_nodes_from(technological_advancements, type="advancement")

# Adding nodes for historical events
historical_events = ["Factory Act", "Luddite Movement", "Great Exhibition"]
G.add_nodes_from(historical_events, type="event")

# Adding nodes for notable individuals
notable_individuals = ["James Watt", "Ada Lovelace", "John Smith"]
G.add_nodes_from(notable_individuals, type="person")

# Adding edges for technological advancements and showcases
advancement_showcase_edges = [
    ("Steam Engine", "Great Exhibition"),
    ("Spinning Jenny", "Industrial Exposition")
]
G.add_edges_from(advancement_showcase_edges, relation="showcase")

# Adding edges for technological impact on historical events
tech_event_edges = [
    ("Steam Engine", "Factory Act"),
    ("Spinning Jenny", "Luddite Movement")
]
G.add_edges_from(tech_event_edges, relation="impact")

# Adding edges for individuals' involvement in events
person_event_edges = [
    ("James Watt", "Great Exhibition"),
    ("Ada Lovelace", "Telegraph"),
    ("John Smith", "Luddite Movement")
]
G.add_edges_from(person_event_edges, relation="involvement")

# Visualization
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=8)
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
